Можно построить плотность распределения для $df['price']$ и сравнить его, например, с экспоненциальным распределением. 

```python
x = df['price']
sp =np.linspace(0, 40000, 35000)
pdf = stats.expon(scale=x.var()**0.5, loc = df['price'].min()).pdf(sp)
plt.figure(figsize=(10,5))
plt.hist(x, density=True, bins=20)
plt.plot(sp, pdf)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

Видно, что они неплохо соотносятся.

Проверим это на qq-plot. 

```python
res = stats.probplot(x, plot=plt, dist='expon')
```

Можно сказать с достаточной точностью, что цены распределены по экспоненциальному закону.

6. С использованием t-test статистики проверить взаимосвязи между несколькими переменными

t-test позволяет сравнивать матожидания одной выборки с каким-либо конкретным числом или же с маотжиданием другой выборки.

Проверим гипотезу, например, $H_0: M_{price} = 13000$ 

```python
stats.ttest_1samp(df['price'], 13000)
```

Так как p-value > 0.05, можно не отвергать данную гипотезу.

```python
df['make'].unique()
```

Проверим гипотезу, например, $H_0: M_{price\ jaguar} = M_{price\ bmw}$ 

```python
df1 = df[df['make']=='jaguar']['price']
```

```python
df2 = df[df['make']=='bmw']['price']
```

```python
stats.ttest_ind(df1, df2)
```

Так как p-value > 0.05, можно не отвергать данную гипотезу. 

Проверим гипотезу о том, что расход топлива в городе и загородом в среднем одинаковый: $H_0: M_{city\ mpg} = M_{highway\ mpg}$ 

```python
stats.ttest_ind(df['city-mpg'], df['highway-mpg'])
```

Получается, что р-значение меньше 0.05 - можно отвергнуть данную гипотезу. 

```python
 df['city-mpg'].mean(), df['highway-mpg'].mean()
```

Видно, что они действительно отличаются.

Проверим гипотезу о том, что цена машины с формой convertible и с формой hardtop одинаковая: $H_0: M_{convertible\ price} = M_{hardtop\ price}$ 

```python
df1 = df[df['body-style']=='convertible']['price']
df2 = df[df['body-style']=='hardtop']['price']
```

```python
stats.ttest_ind(df1, df2 )
```

Так как p-value > 0.05, можно не отвергать данную гипотезу.
