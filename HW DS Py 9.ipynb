# Вопрос 1. 
На шахматной доске размера n на n случайно размещают n ладей. Найдите вероятность событий:


    А = {Ладьи не бьют друг друга}
    B = {Ладьи не бьют друг друга и на главной диагонали нет никаких фигур}
    
Количество всех размещений n ладей на доске n на n равно $n^2(n^2-1)...(n^2 - (n - 1)$

Количество всех размещений n ладей, когда они не бьют друг друга: $n^2(n^2-1)...2^2$

Тогда $P(A) = \frac{(n^2-1)*...*2^2}{(n^2-1)(n^2-2)...(n^2 - (n - 1))}$

Вероятность того, что на главной диагонали нет фигур равна: $P(B') = \frac{n^2 - n}{n^2}$

Тогда $P(B) = \frac{(n^2-n)(n^2-1)*...*2^2}{n^2(n^2-1)(n^2-2)...(n^2 - (n - 1))}$

# Вопрос 2. 
Брошено 3 игральных кости. Найти вероятность того, что на всех костях выпала "шестерка" при условии, что:

    (A) по крайней мере на одной кости выпала "шестерка"
    (A') по крайней мере на двух костях выпало равное количество очков
    
Cобытие B: на всех трех шестерка.

Для решения этой задачи используем формулу условной вероятности. 

$P(A|B) = \frac{P(AB)}{P(B)}$

1) Событие А означает, что благоприятные исходы - это исходы, когда на одной, двух и трех костях выпала 6. То есть можно найти ее если вычесть из 1 вероятность события, когда не на одной кости нет 6.

$P(A) = 1 - \frac{5^3}{6^3} = \frac{91}{216}$

$P(B) = \frac{1}{6^3} = \frac{1}{216}$

Так событие В вложено в событие А, то $P(AB) = P(B)$

Тогда событие В при условии А: $P(B|A) = \frac{P(B)}{P(A)} = \frac{1}{91}$

2) Событие А' означает, что на двух костях выпало одинаковое число, либо на трех. 

То есть вероятность события А' можно искать как сумму вероястностей этих двух возможностей:

$P(A') = \frac{6*5*3}{216} + \frac{6}{216} = \frac{96}{216}$

Так событие В вложено в событие А', то $P(A'B) = P(B)$

Тогда событие В при условии А': $P(B|A') = \frac{P(B)}{P(A')} = \frac{1}{96}$

# Вопрос 3.
В некоторой популяции до 60 лет доживает 50%, а до 80 лет 20%. Какова вероятность, что случайно выбранный шестидесятилетний представитель популяции доживут до восьмидесяти?

А - {дожить до 60}

B - {дожить до 80}

В данном случае можем использовать формулу Байеса: $P(B|A) = \frac{P(A|B)P(B)}{P(A)}$

P(A|B) = 1, так как любой человек, который дожил до 80, уже дожил до 60.

$P(B) = \frac{1}{5}$, $P(A) = \frac{1}{2}$

Тогда $P(B|A) = \frac{2}{5}$

# Вопрос 4.
На заводе обуви производится 3 типа моделей. Вероятность того, что первая модель некачественная 0.1, вторая 0.2 и третья 0.15.
Соотношение моделей на складе 40 : 30 : 30.

1) Какая вероятность того, что случайно выбранная пара обуви окажется качественной?

2) Какая вероятность того, что выбранная некачественная пара обуви окажется второй модели?

В данном случае мы воспользуемся формулой полной вероятности. $\mathbb{P}(A) = \sum\limits_{i=1}^{n} \mathbb{P}( A \mid B_i) \mathbb{P}(B_i)$

A - обувь качественная.

$B_1$ - модель 1.

$B_2$ - модель 2. 

$B_3$ - модель 3. 


1) Введем гипотезы:

$P(B_1) = 0.4$, $P(A|B_1) = 0,9$

$P(B_2) = 0.3$, $P(A|B_2) = 0,8$

$P(B_3) = 0.3$, $P(A|B_3) = 0.85$

Тогда вероятность того, что случайно выбранная пара обуви окажется качественной: $P(A) = 0.4*0.9 + 0.3*0.8 + 0.3*0.85 = 0.855$

2) Тогда вероятность того, что случайно выбранная пара обуви окажется некачественной: $P(A') = 1 - P(A) = 0.145$

Вероятность того, что выбранная некачественная пара обуви окажется второй модели: $P(B_2|A) = \frac{P(B_2)P(A'|B_2)}{P(A')} = \frac{0.3*0.2}{0.145} = 0.4138$

Постройте графики плотностей и функций распределения для:

Для каждого распределения поэкспериментируйте с выбором параметров и зобразите их на одном графике, не забывая про легенду и подписи осей.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

# Вопрос 2.1. Распределение Пуассона.

```python
#для начала рассмотрим распределение Пуассона
#параметр распрделенния Пуассона
mu = 5

# зададим распределенную случайную величину по закону Пуассона
poisson_rv = sts.poisson(mu)
poisson_rv.rvs(size=20)
#сгенерируем 20 значений
```

```python
x = np.linspace(0,10,11)
cdf = poisson_rv.cdf(x)

#Распределение Пуассона - дискретное
plt.figure(figsize=(12,6))
plt.step(x, cdf)
plt.title('Распределение Пуассона')
plt.ylabel('$F(x)$, функция распределения')
plt.xlabel('$x$, значение случайной величины')
```

```python
pmf = poisson_rv.pmf(x)

plt.figure(figsize=(12,6))
plt.plot(x, pmf, 'v')
plt.title('Распределение Пуассона')
plt.ylabel('$P(X=x)$, Функция вероятности')
plt.xlabel('$x$, значение случайной величины')
```

```python
plt.figure(figsize=(12,6))
for mu in [3, 5, 8, 10]:
    rv = sts.poisson(mu)
    cdf = rv.cdf(x)
    plt.step(x, cdf, label="$mu=%s$" % (mu))
plt.legend()
plt.title("Распределение Пуассона для разных параметров mu")

plt.ylabel('$F(X)$, функция распределения')
plt.xlabel('$x$, значение случайной величины')
```

```python
plt.figure(figsize=(12,6))
for mu in [3, 5]:
    rv = sts.poisson(mu)
    pmf = rv.pmf(x)
    plt.plot(x, pmf, 'o', label="$mu=%s$" % (mu))
plt.legend()

plt.title("Распределение Пуассона для разных параметров mu")
plt.ylabel('$f(X)$, Функция вероятности')
plt.xlabel('$x$, значение случайной величины')
```

# Вопрос 2.2. Экспоненциальное распределение.

```python
#Рассмотрим по такому же принципу экспоненциальное распределение, которое уже является непрерывным.
mu = 5 

# зададим распределенную случайную величину по эксп. закону.
expon_rv = sts.expon(scale=mu)

# сгенерируем 20 значений
expon_rv.rvs(size=20)
```

```python
x = np.linspace(0,20,100)
cdf = expon_rv.cdf(x)

#Распределение Пуассона - дискретное
plt.figure(figsize=(12,6))
plt.plot(x, cdf)
plt.title('Функция экспоненциального распределения c параметром lambda = 0.2')
plt.ylabel('$F(x)$, функция распределения')
plt.xlabel('$x$, значение случайной величины')
```

```python
pdf = expon_rv.pdf(x)

plt.figure(figsize=(12,6))
plt.plot(x, pdf)
plt.title('Плотность экспоненциального распределения c параметром lambda = 0.2')
plt.ylabel('$f(x)$, функция плотности распределения')
plt.xlabel('$x$, значение случайной величины')
```

```python
plt.figure(figsize=(12,6))
x = np.linspace(0, 5, 100)
for mu in [2, 1, 2/3]:
    rv = sts.expon(scale=mu)
    cdf = rv.cdf(x)
    plt.plot(x, cdf, label="$lambda=%s$" % (1/mu))

plt.legend()
plt.title("Функция экспоненциального распределения для разных параметров lambda")
plt.ylabel('$F(X)$, функция распределения')
plt.xlabel('$x$, значение случайной величины')
```

```python
plt.figure(figsize=(12,6))
x = np.linspace(0, 5, 100)
for mu in [2, 1, 2/3]:
    rv = sts.expon(scale=mu)
    pdf = rv.pdf(x)
    plt.plot(x, pdf, label="$lambda=%s$" % (1/mu))
plt.legend()

plt.title("Плотность вероятности экспоненциального распределения для разных параметров lambda")
plt.ylabel('$f(X)$, функция плотности распределения')
plt.xlabel('$x$, значение случайной величины')
```

# Вопрос 2.3. Гамма-распределение.

```python
#Рассмотрим еще одно распределение по такому принципу, например гамма-распределение.
a = 1
lambda_par = 0.5

gamma_rv = sts.gamma(a, loc=0, scale=1/lambda_par)

# сгенерируем 30 значений
gamma_rv.rvs(size=30)
```

```python
x = np.linspace(0,10,100)
cdf = gamma_rv.cdf(x)

#Распределение Пуассона - дискретное
plt.figure(figsize=(12,6))
plt.plot(x, cdf)
plt.title('Функция гамма-распределения c параметром lambda = 0.5 и alpha = 1')
plt.ylabel('$F(x)$, функция распределения')
plt.xlabel('$x$, значение случайной величины')
```

```python
pdf = gamma_rv.pdf(x)

plt.figure(figsize=(12,6))
plt.plot(x, pdf)
plt.title(' Плотность гамма-распределения c параметром lambda = 0.5 и alpha = 1')
plt.ylabel('$f(x)$, функция плотности распределения')
plt.xlabel('$x$, значение случайной величины')
```

```python
plt.figure(figsize=(12,6))
x = np.linspace(0, 5, 100)
for a in [2, 1, 7]:
    for lambda_par in [3, 2]:
        rv = sts.gamma(a, loc=0, scale=1/lambda_par)
        cdf = rv.cdf(x)
        plt.plot(x, cdf, label="$lambda=%s, alpha=%s$" % (lambda_par,a))

plt.legend()
plt.title("Функция гамма-распределения для разных параметров lambda и alpha")
plt.ylabel('$F(X)$, функция распределения')
plt.xlabel('$x$, значение случайной величины')
```

```python
plt.figure(figsize=(12,6))
x = np.linspace(0, 5, 100)
for a in [2, 1, 7]:
    for lambda_par in [3, 2]:
        rv = sts.gamma(a, loc=0, scale=1/lambda_par)
        pdf = rv.pdf(x)
        plt.plot(x, pdf, label="$lambda=%s, alpha=%s$" % (lambda_par,a))

plt.legend()
plt.title("Плотность вероятности гамма-распределения для разных параметров lambda и alpha")
plt.ylabel('$f(X)$, функция плотности распределения')
plt.xlabel('$x$, значение случайной величины')
```
